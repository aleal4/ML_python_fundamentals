```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
x=np.linspace(0,10,100)
y=x**3
```

```python
plt.plot(x,y,'red',linewidth=4,linestyle='--')
plt.xlabel('Time')
plt.ylabel('Number of iterations')
plt.title('Number of Iterations Vs. Time')
plt.show()
```

```python
fig=plt.figure(figsize=(10,5))
plt.subplot(1,3,1)
plt.plot(x,y,'b')
plt.title('Plot 1')
plt.subplot(1,3,2)
plt.plot(y*2,x**2,'r')
plt.title('Plot 2')
plt.subplot(1,3,3)
plt.plot(x,y,'purple')
plt.title('Plot 3')
plt.show()
```
