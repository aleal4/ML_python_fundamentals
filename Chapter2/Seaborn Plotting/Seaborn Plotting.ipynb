```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
%matplotlib inline
```

```python
df_Shale= pd.read_csv('Chapter2_Shale_Gas_Wells_DataSet.csv')
```

```python
df_Shale.columns
```

```python
# Create the figure
plt.figure(figsize=(5, 5))

# Plot the histogram
sns.displot(data=df_Shale, x='Well Spacing', color='blue', bins='auto', kde=True)

# Add labels and title
plt.xlabel('Well Spacing')
plt.ylabel('Frequency')
plt.title('Distribution Plot for Well Spacing')

# Show the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats

# Create the jointplot
g = sns.jointplot(
    data=df_Shale,
    x='Stage Spacing',
    y='EUR',
    kind='scatter',
    color='blue'
)

# Calculate Pearson correlation
r, p = stats.pearsonr(df_Shale['Stage Spacing'], df_Shale['EUR'])

# Add correlation text to the plot
plt.annotate(f"Pearson r = {r:.2f}, p = {p:.2g}", xy=(0.05, 0.95), xycoords='axes fraction', ha='left', va='top', fontsize=10)

plt.show()
```
